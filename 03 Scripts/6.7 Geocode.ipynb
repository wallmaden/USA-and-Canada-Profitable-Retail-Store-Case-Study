```python
import pandas as pd
import requests
import os
from time import sleep

# OpenCage API Key
API_KEY = "d7a651e95a5c439e92f42bc569fa4007"

# Define File Paths
path = r"C:\Users\wallm"
file_path = os.path.join(path, '01 2025 Advanced Analytics Project', 'Retail_Superstore_Project', '02 Data', 'cleaned_retail_superstore.csv')

# Load dataset
df_super = pd.read_csv(file_path)

# Step 1️⃣: Get unique cities
unique_cities = df_super['City'].unique()
print(f"✅ Found {len(unique_cities)} unique cities.")

# Step 2️⃣: Function to get coordinates from OpenCage API
def get_coordinates(city):
    sleep(1)  # Prevent rate limiting
    url = f"https://api.opencagedata.com/geocode/v1/json?q={city}&key={API_KEY}"
    
    try:
        response = requests.get(url).json()
        if response.get('results'):
            lat = response['results'][0]['geometry']['lat']
            lon = response['results'][0]['geometry']['lng']
            return lat, lon
    except Exception as e:
        print(f"❌ Error with {city}: {e}")

    return None, None  # Return None if no coordinates found

# Step 3️⃣: Get coordinates for unique cities & store them in a dictionary
city_coords = {}

for city in unique_cities:
    city_coords[city] = get_coordinates(city)
    print(f"📍 Processed city: {city} → {city_coords[city]}")

# Step 4️⃣: Apply saved coordinates to the full dataframe
df_super['Latitude'] = df_super['City'].map(lambda x: city_coords[x][0])
df_super['Longitude'] = df_super['City'].map(lambda x: city_coords[x][1])

# Step 5️⃣: Save the updated file
save_path = os.path.join(path, '01 2025 Advanced Analytics Project', 'Retail_Superstore_Project', '02 Data', 'cleaned_retail_superstore_geo.csv')
df_super.to_csv(save_path, index=False)

print("🎉 Geocoding complete! Latitude & Longitude added successfully!")
```
